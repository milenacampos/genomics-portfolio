```python
# Remove the pedigree-only rows (anoNt == 0)
df_animals = df[df['anoNt'] > 0].copy()
print(f"Animals with birth year: {len(df_animals):,}")
print(f"Removed {len(df) - len(df_animals):,} pedigree-only rows")

# Animals per year after cleaning
df_animals.groupby('anoNt')['animal'].count().sort_index().tail(15)
```

```python
# Define reasonable biological limits for Nellore
limits = {
    'peson':  (5, 60),      # birth weight kg
    'pesod':  (60, 400),    # weaning weight kg  
    'pesos':  (100, 600),   # yearling weight kg
}

for trait, (low, high) in limits.items():
    total = df_animals[trait].notna().sum()
    outliers = df_animals[
        (df_animals[trait] < low) | (df_animals[trait] > high)
    ][trait].count()
    pct = outliers / total * 100
    print(f"{trait}: {outliers:,} outliers ({pct:.1f}%) outside [{low}, {high}] kg")
```

```python
df_clean = df_animals.copy()

for trait, (low, high) in limits.items():
    before = df_clean[trait].notna().sum()
    df_clean.loc[
        (df_clean[trait] < low) | (df_clean[trait] > high), trait
    ] = None
    after = df_clean[trait].notna().sum()
    print(f"{trait}: removed {before - after:,} outliers")

print(f"\nFinal dataset: {len(df_clean):,} animals")
```

```python
def clean_phenotypes(filepath, limits, output_path):
    """
    Load, validate, and clean a Nellore phenotype file.
    
    Args:
        filepath: path to raw CSV file
        limits: dict of {trait: (min, max)} biological limits
        output_path: where to save the clean CSV
    
    Returns:
        df_clean: cleaned DataFrame
    """
    print("=" * 50)
    print("PHENOTYPE CLEANING PIPELINE")
    print("=" * 50)
    
    # Step 1: Load
    print("\n[1/4] Loading data...")
    df = pd.read_csv(filepath, low_memory=False)
    print(f"      Loaded: {len(df):,} rows x {df.shape[1]} columns")
    
    # Step 2: Filter pedigree-only rows
    print("\n[2/4] Filtering pedigree-only rows...")
    df_animals = df[df['anoNt'] > 0].copy()
    removed = len(df) - len(df_animals)
    print(f"      Kept: {len(df_animals):,} animals ({removed:,} pedigree rows removed)")
    
    # Step 3: Remove outliers
    print("\n[3/4] Removing biological outliers...")
    df_clean = df_animals.copy()
    total_outliers = 0
    for trait, (low, high) in limits.items():
        before = df_clean[trait].notna().sum()
        df_clean.loc[
            (df_clean[trait] < low) | (df_clean[trait] > high), trait
        ] = None
        after = df_clean[trait].notna().sum()
        n = before - after
        total_outliers += n
        print(f"      {trait}: {n:,} outliers removed (limits: {low}–{high} kg)")
    print(f"      Total outliers removed: {total_outliers:,}")
    
    # Step 4: Summary report
    print("\n[4/4] Summary report...")
    for trait in limits.keys():
        n = df_clean[trait].notna().sum()
        mean = df_clean[trait].mean()
        sd = df_clean[trait].std()
        print(f"      {trait}: n={n:,}  mean={mean:.1f}  sd={sd:.1f}")
    
    # Save
    df_clean.to_csv(output_path, index=False)
    print(f"\nClean file saved to: {output_path}")
    print("=" * 50)
    
    return df_clean
```

```python
limits = {
    'peson': (5, 60),
    'pesod': (60, 400),
    'pesos': (100, 600),
}

df_clean = clean_phenotypes(
    filepath='/Users/milenacampos/Documents/projeto doc/DEFEITOS NELORE/pedigree+fenotipo.csv',
    limits=limits,
    output_path='/Users/milenacampos/Documents/projeto doc/DEFEITOS NELORE/pedigree+fenotipo_clean.csv'
)
```

```python
# Simple example to understand the concept
import pandas as pd

# Table A — phenotypes for 4 animals
pheno = pd.DataFrame({
    'animal': ['A101', 'A102', 'A103', 'A104'],
    'weight': [180, 210, 195, 220]
})

# Table B — genotypes for 3 of those animals
geno = pd.DataFrame({
    'animal': ['A101', 'A102', 'A103'],
    'SNP1': [0, 1, 2],
    'SNP2': [1, 1, 0]
})

print("Phenotypes:")
print(pheno)
print("\nGenotypes:")
print(geno)
```

```python
inner = pd.merge(pheno, geno, on='animal', how='inner')
print("INNER JOIN:", len(inner), "animals")
print(inner)
```

```python
left = pd.merge(pheno, geno, on='animal', how='left')
print("\nLEFT JOIN:", len(left), "animals")
print(left)
```

```python
outer = pd.merge(pheno, geno, on='animal', how='outer')
print("\nOUTER JOIN:", len(outer), "animals")
print(outer)
```
